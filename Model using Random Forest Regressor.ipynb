#### Code (15 marks)

```python
## Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv("rent_1.csv")
```

```python
# Saving only the integer and float values in the dataset in the variable named "data"

data= pd.read_csv("rent_1.csv",usecols=['bathrooms','bedrooms','latitude','longitude','interest_level','price'])
```

```python
# Coverting the dataset into dataframe.

df = pd.DataFrame(data)
```

```python
# Looking at the top 5 rows of the dataframe

df.head()
```

### Now, whole dataset will be split into two parts:
- Training part : For our model training.
- Test part : For model evaluation.

```python
## Assigning the training data in the variable called 'X'. Since, price needs to be predicted so we will not keep that in X.

X = df.drop(['price'],axis=1)
```

```python
## Assigning price in variable 'Y' to train the model so that it predict correct price based on the data given in X variale.

y= df['price']
```

```python
X.head()
```

```python
y.head()
```

```python
## Importing train_test split library from sklearn.model package
## Splitting X and Y as train and test with test_size= 30%

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## Checking how many rows and columns have randomly been taken for training and testing the model. Since, Y has only 1 column, price
## for training and testing therefore, it only shows the number of rows in the shape.

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## Now, fit the Decision tree Regressor on our training set.
```

```python
## Importing Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
```

```python
## Training model using random forest regressor. Where, number of decision trees would be 200 and as we want OOB score so we have done True for it.

rfr = RandomForestRegressor(n_estimators=200,random_state=0,oob_score = True)
```

```python
rfr.fit(X_train,y_train)
```

```python
# Evaluate the model by using performance metric (R2-Score)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
## Calculating R2 score and Mean Absolute Error for the models we have tested after giving training

print("R2 Score is:",np.round(r2_score(y_test,rfr.predict(X_test))*100,2))
print("MAE:",mean_absolute_error(y_test,rfr.predict(X_test).round(3)))
```

```python
print('Score: ', (rfr.score(X_train, y_train)*100).round(2))
```

```python
print('OOB Score:',(rfr.oob_score_*100).round(3))
```

### We can see that OOB sore is negative that means our model is worse than the best constant predictor.

- Next, RandomforestRegressor has been applied on training set without any data mining/cleaning and different scores calculated by using different metrics.

- Model is giving ~70% accuracy score and R2 score but OOB score is negative(~40%) that means our model is perfroming very poor on unseen data.

### Part 2 - Denoising the data

This section should include the code necessary to **denoise** the data, and it should include what is needed to identify inconsistencies, problems, errors, etc. in the data. 


```python
## Loading the uncleaned data again with all the columns and rows

data= pd.read_csv("rent_1.csv")
```

```python
data.head()
```

```python
## Removing unwanted columns to clean the data

clean_data= data.drop(['building_id', 'created', 'Unnamed: 0', 'description', 'display_address', 'manager_id', 'photos', 'street_address'],axis=1)
clean_data
```

```python
## Now, let's see if there are any null values in the dataset

clean_data.isnull().sum()
```

```python
a = clean_data[clean_data['price'] <= 0]
```

```python
## Deleting the rows where price is either less than 0 or equal to 0

clean_data = clean_data.drop(a.index, axis=0)
```

```python
## Counting number of features in each house and giving the count corresponding to it. Eg: 1st have 3 features, 2nd has 11.

clean_data['features']=clean_data['features'].apply(lambda x : len(x[1:-1].split(",")))
```

```python
clean_data['features']
```

```python
# Replacing blank values with mean i.e. 5

clean_data['features']=clean_data['features'].replace(to_replace='[]',value=5)
```

```python
## Checking if there is still any blank value left in the dataset

clean_data[clean_data['features'] == '[]'].count()
```

```python
clean_data.head()
```

```python
clean_data.describe()
```

```python
clean_data.head()
```

```python
## Considering only those where price lies between 1000 to 10000

clean_data = clean_data[(clean_data['price'] > 1000) & (clean_data['price'] < 10000)]
```

```python
clean_data.shape
```

```python
clean_data[clean_data['latitude'] < 40]
```

```python
 l = clean_data[clean_data['latitude'] ==0]
```

```python
l.index
```

```python
## Dropping those rows where latitue is 0 because they are not useful to us

clean_data=clean_data.drop(l.index,axis=0)
```

```python
clean_data = clean_data[(clean_data['latitude'] !=0) | (clean_data['longitude']) ]
```

```python
clean_data.shape
```

### Part 3 - Creating and evaluating a final model


```python
X = clean_data.drop(['price'],axis=1)
X.head()
```

```python
y=clean_data['price']
```

```python
y.head()
```

```python
## Training the model and testing with size 30%

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=132)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## calculating R squared score

print("R2_Score1:",r2_score(y_test,rfr1.predict(X_test))*100)
```

```python
## calculating Mean Absolute Error

print("MAE1:",mean_absolute_error(y_test,rfr1.predict(X_test).round(3)))
```
